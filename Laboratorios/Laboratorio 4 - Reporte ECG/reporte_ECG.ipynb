## ‚ö°Ô∏è Tipos de Ondas del Electrocardiograma (ECG) [2]

Ya vimos que b√°sicamente el **electrocardiograma (ECG)** es un registro de la **electricidad del coraz√≥n**, donde en el papel o en la pantalla, se ve como una serie de **ondas** que suben y bajan. Cada onda tiene un nombre y nos cuenta algo distinto del coraz√≥n.  

### üìà Ondas principales

<details>
<summary>üü¢ Onda P</summary>

- Es la **primera onda** del ciclo.  
- Representa cuando las **aur√≠culas** (las cavidades superiores del coraz√≥n) se activan.  
- Dura muy poquito, menos de **0.10 segundos**.  
- Si hay problemas en las aur√≠culas, esta onda puede cambiar de forma.  

</details>

<details>
<summary>üîµ Complejo QRS</summary>

- Es un grupo de ondas (Q, R y S).  
- Representa cuando los **ventr√≠culos** (las cavidades grandes del coraz√≥n) se activan.  
- Dura entre **0.06 y 0.10 segundos**.  
- La onda **Q** es negativa, la **R** es positiva y la **S** vuelve a ser negativa.  
- Si todo el complejo es negativo, se llama **QS**.  

</details>

<details>
<summary>üü† Onda T</summary>

- Representa cuando los ventr√≠culos se **recuperan** (repolarizaci√≥n).  
- Normalmente es **positiva**, excepto en algunas derivaciones.  
- Su forma es asim√©trica: sube lento y baja r√°pido.  

</details>

<details>
<summary>üü£ Onda U</summary>

- Es una onda peque√±ita que aparece **despu√©s de la T**.  
- No siempre se ve y los m√©dicos todav√≠a no est√°n seguros de qu√© significa.  
- Puede estar relacionada con los **m√∫sculos papilares** del coraz√≥n.  
- Se nota m√°s en ciertos problemas como la **falta de potasio**.  

</details>

### üß† Recuerda
- El ECG es como un **idioma del coraz√≥n**: cada onda nos cuenta una parte de su historia.  
- Si las ondas cambian de forma, pueden indicar un problema.  
- Por eso, los m√©dicos lo usan para diagnosticar enfermedades del coraz√≥n.  


# Reposo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar archivo
signal = np.loadtxt("Data_ECG/Reposo.txt")
# Seleccionar la columna 5 (para poder eliminar valores peque√±os o ceros)
reposo = np.loadtxt("Data_ECG/Reposo.txt", comments="#")[:,5]

fs = 1000  # frecuencia de muestreo
t = np.arange(len(signal)) / fs

# Mostrar 5 segundos (5000 muestras)
samples = 5000
# Graficar
plt.figure(figsize=(12,5))
plt.plot(t[:samples], reposo[:samples], color="blue")

plt.title("Se√±al ECG - Se√±al en Reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]", fontsize=12)
plt.ylabel("Amplitud [¬µV]", fontsize=12)

# Acotar el eje Y 
plt.ylim(0, 1000)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
reposo = np.loadtxt("Data_ECG/Reposo.txt", comments="#")[:,5]
contener = np.loadtxt("Data_ECG/contener aire1.txt", comments="#")[:,5]

fs = 1000
t = np.arange(len(reposo)) / fs

# Mostrar 10 segundos (10000 muestras)
samples = 10000

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(t[:samples], reposo[:samples], color="blue")
plt.title("Se√±al ECG - Se√±al en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [¬µV]")
plt.ylim(0,1000); plt.grid(True, linestyle="--", alpha=0.6)

plt.subplot(2,1,2)
plt.plot(t[:samples], contener[:samples], color="red")
plt.title("Se√±al ECG - Contener Respiraci√≥n N¬∞1", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [¬µV]")
plt.ylim(0,1000); plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
