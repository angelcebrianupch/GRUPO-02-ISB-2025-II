# Reposo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar archivo
signal = np.loadtxt("Data_ECG/Reposo.txt")
# Seleccionar la columna 5 (para poder eliminar valores pequeos o ceros)
reposo = np.loadtxt("Data_ECG/Reposo.txt", comments="#")[:,5]

fs = 1000  # frecuencia de muestreo
t = np.arange(len(signal)) / fs

# Mostrar 5 segundos (5000 muestras)
samples = 5000
# Graficar
plt.figure(figsize=(12,5))
plt.plot(t[:samples], reposo[:samples], color="blue")

plt.title("Seal ECG - Seal en Reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]", fontsize=12)
plt.ylabel("Amplitud [V]", fontsize=12)

# Acotar el eje Y 
plt.ylim(0, 1000)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

# Contener respiracin N1

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
reposo = np.loadtxt("Data_ECG/Reposo.txt", comments="#")[:,5]
contener = np.loadtxt("Data_ECG/contener aire1.txt", comments="#")[:,5]

fs = 1000
t = np.arange(len(reposo)) / fs

# Mostrar 10 segundos (10000 muestras)
samples = 10000

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(t[:samples], reposo[:samples], color="blue")
plt.title("Seal ECG - Seal en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [V]")
plt.ylim(0,1000); plt.grid(True, linestyle="--", alpha=0.6)

plt.subplot(2,1,2)
plt.plot(t[:samples], contener[:samples], color="red")
plt.title("Seal ECG - Contener Respiracin N1", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [V]")
plt.ylim(0,1000); plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
