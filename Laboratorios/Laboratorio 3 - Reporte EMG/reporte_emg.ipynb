## Bíceps en reposo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar archivo
data = np.loadtxt("Data_EMG/biceps reposo.txt")

# Seleccionar la columna 5 (para poder eliminar valores pequeños o ceros)
signal = data[:, 5]

fs = 2000  # frecuencia de muestreo
t = np.arange(len(signal)) / fs

# Graficar
plt.figure(figsize=(12,5))
plt.plot(t[:2000], signal[:2000], color="navy", linewidth=1)

plt.title("Señal EMG - Bíceps en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]", fontsize=12)
plt.ylabel("Amplitud [µV]", fontsize=12)

# Acotar el eje Y 
plt.ylim(400, 700)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
