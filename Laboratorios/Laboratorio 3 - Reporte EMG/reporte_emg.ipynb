# üìö INTRODUCCI√ìN

La electromiograf√≠a (EMG) es una t√©cnica utilizada para registrar la actividad el√©ctrica generada por los m√∫sculos esquel√©ticos durante el reposo y la contracci√≥n. El an√°lisis de se√±ales EMG permite estudiar el funcionamiento neuromuscular, identificar patrones de activaci√≥n muscular y evaluar la respuesta ante diferentes tipos de movimientos y esfuerzos. En este laboratorio, se registraron se√±ales EMG de distintos m√∫sculos (b√≠ceps, m√∫sculos faciales y abductor corto del pulgar) bajo condiciones de reposo, movimientos voluntarios y contrafuerza, con el objetivo de comparar la actividad el√©ctrica en cada situaci√≥n y comprender la fisiolog√≠a subyacente.

# üîé OBJETIVOS

- Registrar y analizar se√±ales EMG de diferentes grupos musculares en condiciones de reposo, movimiento voluntario y contrafuerza.
- Comparar la actividad el√©ctrica muscular entre las distintas condiciones y movimientos.
- Visualizar y describir las diferencias en la se√±al EMG seg√∫n el tipo de m√∫sculo y el esfuerzo realizado.
- Comprender la relaci√≥n entre la se√±al EMG y la activaci√≥n muscular durante distintas tareas motoras.

## 1.1. Ubicaci√≥n de electrodos

### ‚úÖ Materiales
- 2 electrodos desechables (activos)  
- 1 electrodo de referencia/masa (tierra)  
- Toallitas con alcohol o algod√≥n con alcohol  
- Tijera peque√±a si hay mucho vello
- BITalino y laptop con **OpenSignals**  

### üõ°Ô∏è Antes de empezar
- El EMG **solo registra** la electricidad del m√∫sculo, no introduce corriente.  
- No usar electrodos sobre heridas o piel irritada.  
- Los electrodos son **desechables** para higiene y buena calidad de se√±al.  

### üîé Paso a paso

1. **Coloca a la persona**  
   - Sentada, con la espalda recta.  
   - El brazo relajado o con el codo a 90¬∞, palma hacia arriba.  

2. **Localiza la l√≠nea gu√≠a**  
   - Imagina una l√≠nea desde el **hombro (acromion)** hasta el **pliegue del codo (fosa cubital)**.  

3. **Marca la posici√≥n central**  
   - Encuentra el punto que est√° a **1/3 de la distancia desde el codo hacia el hombro**.  
   - Ese es el centro del par de electrodos.  

4. **Prepara la piel**  
   - Recorta el vello si es necesario.  
   - Limpia con alcohol y deja secar.  

5. **Coloca los 2 electrodos activos**  
   - P√©galos en el vientre del b√≠ceps, **alineados con las fibras musculares**.  
   - Mant√©n una separaci√≥n de **‚âà20 mm (2 cm)** entre los centros.  

6. **Coloca el electrodo de referencia (tierra)**  
   - Ponlo en una zona √≥sea o sin m√∫sculo cercano, para este experimento se us√≥ la mu√±eca.  

7. **Prueba la se√±al**  
   - Pide que la persona flexione el codo.  
   - En **OpenSignals** ver√°s c√≥mo la se√±al aumenta cuando el b√≠ceps se activa.  

8. **Evita errores comunes**  
   - No pegar los electrodos sobre tendones.  
   - No separar demasiado los electrodos (para no captar otros m√∫sculos).

## 2.1. Ubicaci√≥n de electrodos

- 2 electrodos superficiales desechables (activos)  
- 1 electrodo de referencia/masa (tierra)  
- Toallitas con alcohol o algod√≥n con alcohol  
- Cinta m√©dica para fijar cables

### üõ°Ô∏è Antes de empezar
- El EMG **no duele** y solo registra la electricidad del m√∫sculo.  
- Usa electrodos desechables por higiene.  
- Evita colocarlos sobre granitos, heridas o piel irritada.  
- Pide a la persona que est√© tranquila y sin moverse mucho.  

### üîé Paso a paso (ejemplo: m√∫sculo masetero o cigom√°tico)

1. **Coloca a la persona**  
   - Sentada, con la cabeza relajada y mirando al frente.  

2. **Elige el m√∫sculo de la cara**  
   - **Masetero:** m√∫sculo de la mand√≠bula, se activa al apretar los dientes.  
   - **Cigom√°tico mayor:** m√∫sculo de la mejilla, se activa al sonre√≠r.  

3. **Prepara la piel**  
   - Limpia suavemente con una toallita con alcohol.  
   - Deja que se seque antes de colocar los electrodos.  

4. **Coloca los 2 electrodos activos**  
   - P√©galos **paralelos a las fibras del m√∫sculo elegido**.  
   - Mant√©n una separaci√≥n peque√±a: **‚âà10‚Äì15 mm** (1 a 1.5 cm).  
   - Ejemplo:  
     - Masetero ‚Üí ponlos en la parte lateral de la mand√≠bula, entre el p√≥mulo y el √°ngulo de la quijada.  
     - Cigom√°tico ‚Üí ponlos en la mejilla, desde el p√≥mulo hacia la comisura de la boca.  

5. **Coloca el electrodo de referencia (tierra)**  
   - Ponlo en una zona neutra sin mucha actividad muscular, por ejemplo detr√°s de la oreja.  

6. **Prueba la se√±al**  
   - Pide a la persona que sonr√≠a (cigom√°tico) o que apriete los dientes (masetero).  
   - En **OpenSignals**, deber√≠as ver c√≥mo la se√±al aumenta durante la contracci√≥n.

# CONCLUSION GENERAL

Podemos concluir que:

- La electromiograf√≠a (EMG) permiti√≥ registrar y analizar la actividad el√©ctrica de diferentes grupos musculares bajo condiciones de reposo, movimiento voluntario y contrafuerza, evidenciando variaciones claras en la amplitud y el patr√≥n de las se√±ales seg√∫n el tipo y la intensidad del movimiento.

- Se observ√≥ que los m√∫sculos presentan una se√±al EMG de baja amplitud en reposo, mientras que durante los movimientos voluntarios y especialmente bajo contrafuerza, la amplitud y la complejidad de la se√±al aumentan significativamente, reflejando una mayor activaci√≥n muscular.

- La comparaci√≥n entre los distintos m√∫sculos (b√≠ceps, m√∫sculos faciales y abductor corto del pulgar) mostr√≥ que cada grupo muscular tiene caracter√≠sticas propias en su se√±al EMG, dependiendo de su funci√≥n y del tipo de movimiento realizado.

- El an√°lisis de las se√±ales EMG es fundamental para comprender la fisiolog√≠a muscular y puede ser aplicado en √°reas como la rehabilitaci√≥n, el control de pr√≥tesis y el desarrollo de interfaces hombre-m√°quina.

- Para el proyecto final del curso, se emplear√° la toma de muestras EMG del abductor corto del pulgar, ya que este m√∫sculo permite registrar movimientos precisos y es relevante para aplicaciones de control y reconocimiento de gestos en dispositivos biom√©dicos.
