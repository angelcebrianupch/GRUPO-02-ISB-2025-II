## Bíceps movimiento voluntario N°1

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
reposo = np.loadtxt("Data_EMG/biceps reposo.txt", comments="#")[:,5]
voluntario = np.loadtxt("Data_EMG/biceps voluntario 1.txt", comments="#")[:,5]

fs = 1000
t = np.arange(len(reposo)) / fs

# Mostrar 10 segundos (10000 muestras)
samples = 10000

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(t[:samples], reposo[:samples], color="blue")
plt.title("Señal EMG - Bíceps en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [µV]")
plt.ylim(400,700); plt.grid(True, linestyle="--", alpha=0.6)

plt.subplot(2,1,2)
plt.plot(t[:samples], voluntario[:samples], color="red")
plt.title("Señal EMG - Bíceps en cmovimiento voluntario N°1", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud [µV]")
plt.ylim(400,700); plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()

```
