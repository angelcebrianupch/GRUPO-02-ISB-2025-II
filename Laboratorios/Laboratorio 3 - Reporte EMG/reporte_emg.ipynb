## Movimiento - Bíceps

## Bíceps en reposo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar archivo
signal = np.loadtxt("Data_EMG/biceps reposo.txt")
# Seleccionar la columna 5 (para poder eliminar valores pequeños o ceros)
reposo = np.loadtxt("Data_EMG/biceps reposo.txt", comments="#")[:,5]

fs = 1000  # frecuencia de muestreo
t = np.arange(len(signal)) / fs

# Mostrar 10 segundos (10000 muestras)
samples = 10000
# Graficar
plt.figure(figsize=(12,5))
plt.plot(t[:samples], reposo[:samples], color="blue")

plt.title("Señal EMG - Bíceps en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]", fontsize=12)
plt.ylabel("Amplitud [µV]", fontsize=12)

# Acotar el eje Y 
plt.ylim(400, 700)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
