## Bíceps en reposo

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar archivo
data = np.loadtxt("Data_EMG/biceps reposo.txt")

# Seleccionar la columna 5 (índice 5 en Python → sexta columna)
signal = data[:, 5]

fs = 2000  # frecuencia de muestreo (ajústala si es distinta)
t = np.arange(len(signal)) / fs

# Graficar
plt.figure(figsize=(12,5))
plt.plot(t[:2000], signal[:2000], color="navy", linewidth=1)

plt.title("Señal EMG - Bíceps en reposo", fontsize=14, fontweight="bold")
plt.xlabel("Tiempo [s]", fontsize=12)
plt.ylabel("Amplitud [µV]", fontsize=12)

# Acotar el eje Y para que se vea como tu imagen 2
plt.ylim(400, 700)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
